```python
from functools import lru_cache
@lru_cache(maxsize = None)

def get_earnings_call_url(ticker):
  url = f'https://www.google.com/search?q=ticker%3A{ticker}+Q1+2024+earnings+call+transcript+yahoo+finance'

  res = requests.get(url, headers = {'User-Agent': generate_user_agent()})

  if res.status_code>200:
    return

  soup = BeautifulSoup(res.content, 'html.parser')
  tags = [tag for tag in soup.find_all('a', href = True) if tag.text.lower().find('highlights') == -1]
  hrefs = [tag['href'] for tag in tags if 'finance.yahoo.com' in tag['href']]

  hrefs = ['http'+href.split('http')[-1] for href in hrefs]
  hrefs = [href.split('&')[0] for href in hrefs]


  for earnings_call_url in hrefs[:3]:
    res = requests.get(earnings_call_url, headers={'User-Agent': generate_user_agent()})
    if res.status_code>200:
      print('Error')

    soup = BeautifulSoup(res.content, 'html.parser')

    text = soup.text.lower()
    operator_idx = text.find('operator')
    if operator_idx<0:
      continue
    text = text[operator_idx:]
    return text
```

```python
import spacy

!python -m spacy download en_core_web_md
nlp = spacy.load('en_core_web_md')
```

```python
import nltk
import re
import en_core_web_md
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')

nlp = en_core_web_md.load()
good_pos = ['NOUN', 'VERB', 'ADJ', 'ADV']

mystopwords = set(stopwords.words('english'))

def cleanTextAndLemmatizer(s):
  s = str(s).lower()
  s = re.sub(r'[^\w\s]','',s)
  tokens = word_tokenize(s)

  doc = nlp(' '.join(tokens))

  return [w.lemma_ for w in doc if w.pos_ in good_pos and w.text not in mystopwords]
```

```python
print(cleanTextAndLemmatizer(stock_df['EarningsCall'][:]))
```

```python
stock_df['EarningsCallLemmatized'] = stock_df['EarningsCall'].map(lambda s:' '.join(cleanTextAndLemmatizer(str(s))))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(stock_df['EarningsCallLemmatized'].map(str))
print(X.toarray())
print(vectorizer.get_feature_names_out())
```

```python
from sklearn.model_selection import train_test_split

# obtain indices
X_train, X_test, y_train, y_test, train_idx, test_idx = train_test_split(X, stock_df['y'], stock_df.index, test_size=0.2, random_state=42, stratify=stock_df['y'])
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn import metrics

# train the Logistic Regression model with cross-validation
clf = LogisticRegressionCV(cv=5, random_state=0, class_weight='balanced').fit(X_train, y_train)

# predict the classes
y_pred = clf.predict(X_test)
```

```python
# classification report
print(metrics.classification_report(y_test, y_pred))
```

```python
coefficients = pd.Series(clf.coef_[0], index=vectorizer.get_feature_names_out())

# identify significant words
top_negative_coefficients = coefficients.sort_values().head(20)
top_positive_coefficients = coefficients.sort_values().tail(20)

print("Top Negative Coefficients:\n", top_negative_coefficients)
print("\nTop Positive Coefficients:\n", top_positive_coefficients)
```

```python
import gensim
from gensim.test.utils import common_texts
from gensim.corpora.dictionary import Dictionary
from gensim.models.ldamodel import LdaModel
import matplotlib.pyplot as plt
```

```python
common_texts = stock_df['EarningsCallLemmatized'].map(str).str.split().tolist()

common_dictionary = Dictionary(common_texts)
common_dictionary.filter_extremes(no_above=0.25)
common_corpus = [common_dictionary.doc2bow(text) for text in common_texts]
```

```python
num_topics = 10
lda = LdaModel(common_corpus, num_topics = num_topics, id2word = common_dictionary, passes = 30)
```

```python
for i in range(num_topics):
  plt.figure()
  wordcloud = WordCloud(background_color = 'white').fit_words(dict(lda.show_topic(i, 20)))
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.title(f"Topic #{i}")
  plt.show()
  print()
```

```python
# ARKK and Non-ARKK Stocks
arkk_indices = stock_df[stock_df['y'] == 1].index
non_arkk_indices = stock_df[stock_df['y'] == 0].index
```

```python
# length of common_corpus
print(f"Length of common_corpus: {len(common_corpus)}")

# first few indices of non_arkk_indices to ensure they are within range
print(f"First few non_arkk_indices: {non_arkk_indices[:10]}")
print(f"Max index in non_arkk_indices: {max(non_arkk_indices)}")
```

```python
non_arkk_indices = [i for i in non_arkk_indices if i < len(common_corpus)]
```

```python
# topic distributions for ARKK and non-ARKK stocks
arkk_corpus = [common_corpus[i] for i in arkk_indices]
arkk_topics = [lda[doc] for doc in arkk_corpus]

non_arkk_corpus = [common_corpus[i] for i in non_arkk_indices]
non_arkk_topics = [lda[doc] for doc in non_arkk_corpus]
```

```python
# summarize topics for ARKK and non-ARKK stocks
arkk_topic_distribution = [dict(doc) for doc in arkk_topics]
arkk_topic_summary = pd.DataFrame(arkk_topic_distribution).mean().sort_values(ascending=False)

non_arkk_topic_distribution = [dict(doc) for doc in non_arkk_topics]
non_arkk_topic_summary = pd.DataFrame(non_arkk_topic_distribution).mean().sort_values(ascending=False)
```
